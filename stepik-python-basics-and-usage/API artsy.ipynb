```python
import requests
import json
```

```python
client_id = '3ac4b561e85cf9c605f0'
client_secret = '0db0d5836bc8f31f03275316618b8cab'
```

```python
# инициируем запрос на получение токена
r = requests.post("https://api.artsy.net/api/tokens/xapp_token",
                  data={
                      "client_id": client_id,
                      "client_secret": client_secret
                  })

# разбираем ответ сервера
j = json.loads(r.text)

# достаем токен
token = j["token"]
```

```python
# создаем заголовок, содержащий наш токен
headers = {"X-Xapp-Token" : token}

with open('dataset_24476_4.txt') as f:
    data = f.read().split()

result = []
for item in data:
    # инициируем запрос с заголовком
    r = requests.get(f"https://api.artsy.net/api/artists/{item}", headers=headers)
    r.encoding = 'utf-8'
    # разбираем ответ сервера
    result.append(json.loads(r.text))
```

```python
result = sorted(result, key=lambda x: (x['birthday'], x['sortable_name']))
```

```python
with open('artsy.txt', 'w') as file:
    for artist in result:
        file.write(artist['sortable_name'] + '\n')
```
