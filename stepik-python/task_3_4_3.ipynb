```python
with open('dataset_3363_4.txt') as file:
    result = [line.strip().split(';') for line in file]
    mean_list = [round(sum(map(int, item[1:]))/3, 9) for item in result]
    for mean in mean_list:
        print(mean)
    ovarall_mean_list = []
    ovarall_mean_list.append(round((sum(map(int, [math for name, math, physics, russian in result])) / len(result)), 9))
    ovarall_mean_list.append(round((sum(map(int, [physics for name, math, physics, russian in result])) / len(result)), 9))
    ovarall_mean_list.append(round((sum(map(int, [russian for name, math, physics, russian in result])) / len(result)), 9))
    print(', '.join(map(str, ovarall_mean_list)))
    
with open('output.txt', 'w') as file:
    for mean in mean_list:
        file.write(str(mean))
        file.write("\n")
    file.write(' '.join(map(str, ovarall_mean_list)))
```
